```python
#NYPD failas:
#Su vizualizacija:
# Kurios valandos, kurie paros metai buvo pavojingiausi susišaudymų prasme?
#A) Kuriose valandose mirtinų susišaudymų kiekis buvo didžiausias?
#B) Kuriose valandose procentinis mirtinų susišaudymų kiekis yra didžiausias?
#Ar A ir B valandos sutampa, arne?

import pandas as pd
import matplotlib.pyplot as plt

saudymai = pd.read_csv('NYPD_Shooting.csv', sep=';', parse_dates={'DATA':['OCCUR_DATE','OCCUR_TIME']})

saudymai['HOUR'] = saudymai['DATA'].dt.hour

def paros_metas(hour):
    if hour < 7:
        return 'Naktis'
    elif hour < 12:
        return 'Rytas'
    elif hour < 18:
        return 'Diena'
    else:
        return 'Vakaras'

saudymai['HOUR_GROUP'] = saudymai['HOUR'].apply(paros_metas)

langas, grafikas = plt.subplots(4,1, figsize=(20,20))

all_cases_by_hour = saudymai['HOUR'].value_counts().sort_index()
all_cases_by_hour_group = saudymai['HOUR_GROUP'].value_counts().sort_index()
murders_by_hour = saudymai.query('STATISTICAL_MURDER_FLAG == True')['HOUR'].value_counts().sort_index()
murders_by_hour_group = saudymai.query('STATISTICAL_MURDER_FLAG == True')['HOUR_GROUP'].value_counts().sort_index()
murders_percent_by_hour = round(murders_by_hour/all_cases_by_hour*100)
grafikas[0].plot(all_cases_by_hour.index, all_cases_by_hour.values)
grafikas[0].set_xticks(all_cases_by_hour.index)
grafikas[0].set_xlabel('Valanda')
grafikas[0].set_ylabel('Susišaudymų sk.')
grafikas[0].set_title('Susišaudymai pavalandžiui')
grafikas[1].bar(all_cases_by_hour_group.index, all_cases_by_hour_group.values)
grafikas[1].set_xlabel('Paros metas')
grafikas[1].set_ylabel('Susišaudymų sk.')
grafikas[1].set_title('Susišaudymai pagal paros metą')
grafikas[2].bar(murders_by_hour.index, murders_by_hour.values)
grafikas[2].set_xticks(murders_by_hour.index)
grafikas[2].set_xlabel('Valanda')
grafikas[2].set_ylabel('Mirtinų susišaudymų sk.')
grafikas[2].set_title('Mirtini susišaudymai pavalandžiui')
grafikas[3].plot(murders_percent_by_hour.index, murders_percent_by_hour.values)
grafikas[3].set_xticks(murders_percent_by_hour.index)
grafikas[3].set_xlabel('Valanda')
grafikas[3].set_title('Susišaudymų mirtingumas pavalandžiui')
plt.show()
```

```python
Su life-exp.csv:
life=pd.read_csv('life-exp.csv')
#atrinkite su .query() į naują lentelę tik 2019 metų duomenis.
life_new = life.query('Year == 2019')
life_new.head(5)
# Toliau naudojama tik naujoji lentelė
# Suraskite vidutinę gyvenimo trukmę
vid_gyv_trukmė = life_new['Life expectancy'].mean()
print('vidutinė gyvenimo trukmė', vid_gyv_trukmė)
# Suskirstykite valstybes į ilgaamžes ir ne
# jei gyvenimo trukmė mažiau nei vidurkis - neilgaamžė
# jei didesnė - ilgaamžė.
# Kiek buvo valstybių ilgaamžių, neilgaamžių?
#  .query() duomenų filtravimas ir atrinkimas
def gyv_trukme(amzius):
    if amzius > vid_gyv_trukmė:
        return "ilgaamžis"
    else: return "neilgaamžis"
life_new['Ilgaamžiškumas'] = life_new['Life expectancy'].apply(gyv_trukme)    
ilgaam = life_new[life_new['Life expectancy'] > vid_gyv_trukmė]['Life expectancy'].count()
neilgaam = life_new[life_new['Life expectancy'] < vid_gyv_trukmė]['Life expectancy'].count()
print('Ilgaamžių valstybių kiekis', ilgaam)
print('Neilgaamžių valstybių kiekis', neilgaam)
ilg = life_new["Ilgaamžiškumas"].value_counts()
print(ilg)
```

```python
def year_fix(a):
    if str(a)[:3] == 'XIV':
        return 1301
    elif str(a)[:3] == 'XIX':
        return 1801
    else:
        return int(str(a)[:4])

def amzius(a):
    return 2024 - int(a)

miestai['Miesto amžius'] = miestai['Paminėjimo metai'].apply(year_fix)
miestai['Teisių amžius'] = miestai['Miesto amžius'].apply(amzius)
miestai['Gyventojai'] = miestai['Gyventojų skaičius']

miestai50 = miestai.query("`Gyventojų skaičius` > 50000")
print(miestai50["Gyventojų skaičius"].mean())

# miestai1500 = miestai.query("`Miesto amžius` < 1500")
# print(miestai1500['Gyventojų skaičius'].sum())
print(miestai.query("`Miesto amžius` < 1500")['Gyventojų skaičius'].sum())
```
